```python
import subprocess

def install_git():
    try:
        subprocess.check_call(["sudo", "apt-get", "update"])
        subprocess.check_call(["sudo", "apt-get", "install", "git"])
        print("Git installed successfully!")
    except subprocess.CalledProcessError as e:
        print(f"An error occurred: {e}")

# Run the function
install_git()

```

```python
import subprocess

def install_memote():
    try:
        subprocess.check_call(["pip", "install", "memote"])
        print("Memote installed successfully!")
    except subprocess.CalledProcessError as e:
        print(f"An error occurred: {e}")

# Run the function
install_memote()

```

```python
from google.colab import files

# Prompt user to upload a file
uploaded = files.upload()

# Get the current working directory
current_directory = %pwd

# Access the uploaded file
for filename, content in uploaded.items():
    # Concatenate the current directory and filename to get the full path
    file_path = f"{current_directory}/{filename}"

    with open(file_path, 'wb') as f:
        f.write(content)

    print(f"File uploaded successfully! Path: {file_path}")
```
